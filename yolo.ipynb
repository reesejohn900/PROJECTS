```python
cap = cv2.VideoCapture('C:/Users/pgbos/Downloads/seconds_Animation.mp4')
fps = cap.get(cv2.CAP_PROP_FPS)
if cap.isOpened():
    current_frame = 0
    while True:
        ret, frame = cap.read()
        if ret:
            name = f'C:/Users/pgbos/OneDrive/Desktop/frame_img/{current_frame/fps}.jpg'
            print(f"Creating file... {name}")
            cv2.imwrite(name, frame)
        else:
            break
        current_frame += 1
    cap.release()
cv2.destroyAllWindows()
```
