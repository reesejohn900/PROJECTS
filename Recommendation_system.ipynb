## Recommendation System


* Model 1 : Works by removing duplicates having same **`product_name`** and **`uniq_id`**
* Model 2 : Works by removing duplicates having the same **`product_name`** only

### Model 1 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
## creating a function for creating combined data
def dataset(name):
  data = pd.concat([df_fk[df_fk['product_name']==name],df_amz])
  data.dropna(inplace=True)
  data.drop_duplicates(subset=['product_name','uniq_id'],inplace=True)
  return data
```

```python
## writng a funtion for recommendation

def recommendation(product_name):

  # checking the product name is valid or not
  if product_name not in df_fk['product_name'].to_list():
    print('Product not found!!!!\nEnter a Valid Product Name')

  else:
    data = dataset(product_name)

    #converting text into sparse matrix
    tfidf = TfidfVectorizer(max_features=50000,stop_words='english')
    x = tfidf.fit_transform(data['description'].values.astype('U'))

    # product for which we are getting recommendation
    query = x[0]


    scores = cosine_similarity(query,x)
    scores = scores.flatten()

    ## recommended indexes
    recommended_index = (-scores).argsort()[len(df_fk[df_fk['product_name']==product_name]):
                                            len(df_fk[df_fk['product_name']==product_name])+len(df_fk[df_fk['product_name']==product_name])]

    

    amazon_name = data.iloc[recommended_index]['product_name'].to_list()
    amazon_retail = data.iloc[recommended_index]['retail_price'].to_list()
    amazon_discount = data.iloc[recommended_index]['discounted_price'].to_list()

    ## constructing dataframe for recommended products 
    dict1 = {'Product name in Flipkart':product_name,
            'Retail Price in Flipkart':df_fk[df_fk['product_name']==product_name]['retail_price'],
            'Discounted Price in Flipkart':df_fk[df_fk['product_name']==product_name]['discounted_price'],
            'Product name in Amazon':amazon_name,
            'Retail Price in Amazon':amazon_retail,
            'Discounted Price in Amazon':amazon_discount}


    recommended_data = pd.DataFrame(dict1)
    return recommended_data     
```

```python
recommendation('FabHomeDecor Fabric Double Sofa Bed')
```

```python
recommendation("Alisha Solid Women's Cycling Shorts")
```
